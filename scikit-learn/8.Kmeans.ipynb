```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans # Algoritmo
from sklearn import metrics # Metricas del algoritmo
```

```python
fifa = pd.read_csv('complete.csv') # Lectura del archivo
fifa = pd.DataFrame(fifa) # convertir el archivo en un DataFrame
fifa = fifa.iloc[:,0:94] #Seleccion de las caracteristicas del archivo 
```

```python
ligas =fifa.query('(league == "Spanish Primera División" )|(league=="Colombian Primera A")')
print(ligas)
# Se seleccionan dos ligas para Realizar una comparacion entre los jugadores
```

```python
equipos =fifa.query('(club == "Real Madrid CF" )|(club=="FC Barcelona")|(club=="Atlético Madrid")|(club=="Independiente Medellín")|(club=="Atletico Nacional Medellin")|(club=="Independiente Santa Fe")')
# Se seleccionan 3 equipos de cada liga con toda su nomina completa de jugadores                 
```

```python
equipos = equipos.sort_values(by=['weight_kg'],ascending=[True])# Se ordenan por el peso para eliminar el orden basado en la liga
equipos = equipos.reset_index(drop=True )# se resetea el indice de los datos iniciales
```

```python
# se detecto que  los datos presentan valor NA/NaN (Not a Number)
equipos = equipos.fillna(value=0) # Se procede a reemplazar los NaN por 0
```

```python
jugadores = equipos.iloc[:,68:]
jugadores2 = equipos.iloc[:,2:3]
# Se seleccionan las columnas que van a ser tenidas en cuenta para la comparacion
                     # Todas son de Datos numericos
```

```python
jugadores = jugadores.values # convertir dataframe en array de arrays
```

```python
km = KMeans(n_clusters=4, max_iter=100000)  # # creacion del algoritmo con 4 clusters
```

```python
"""Se plantean las 4 posiciones principales en las que se supone estan los 4 grupos de jugadores
    'portero'
    'defensor'
    'volante'
    'atacante'          """
```

```python
km.fit(jugadores) # Se realiza el ajuste (entrenamiento) del algoritmo con los datos de entrenamiento
```

```python
predicciones = km.predict(jugadores)# lectura de los registros para una prediccion
```

```python
predicciones # grupos de cada una de las predicciones
```
