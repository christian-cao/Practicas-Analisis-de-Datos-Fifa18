```python
import pandas as pd
import sklearn
from sklearn.model_selection import train_test_split 
from sklearn.neural_network import MLPClassifier # Clasificador/ Red Neuronal
```

```python
fifa = pd.read_csv('complete.csv') # Lectura del archivo
fifa = pd.DataFrame(fifa) # convertir el archivo en un DataFrame
fifa = fifa.iloc[:,0:94] #Seleccion de las caracteristicas del archivo 
```

```python
ligas =fifa.query('(league == "Spanish Primera División" )|(league=="Colombian Primera A")')
print(ligas)
# Se seleccionan dos ligas para Realizar una comparacion entre los jugadores
```

```python
equipos =fifa.query('(club == "Real Madrid CF" )|(club=="FC Barcelona")|(club=="Atlético Madrid")|(club=="Independiente Medellín")|(club=="Atletico Nacional Medellin")|(club=="Independiente Santa Fe")')
# Se seleccionan 3 equipos de cada liga con toda su nomina completa de jugadores    
```

```python
equipos = equipos.sort_values(by=['weight_kg'],ascending=[True])# Se ordenan por el peso para eliminar el orden basado en la liga
equipos = equipos.reset_index(drop=True )# se resetea el indice de los datos iniciales
```

```python
# se detecto que  los datos presentan valor NA/NaN (Not a Number)
equipos = equipos.fillna(value=0) # Se procede a reemplazar los NaN por 0
equipos['y']= 0 # se crea una nueva columna que tomara el tipo de clasificacion por cada liga
```

```python
for i, liga in enumerate(equipos['league']):
    if (liga == "Spanish Primera División"):
        equipos['y'][i]= 1
    else:
        equipos['y'][i]= 0

""" Se añadio la columna 'y' (ultima columna ) para identificar a que liga pertenece cada jugador
    0 : Liga colombiana
    1 : Liga Española"""
```

```python
equipos
```

```python
jugadores = equipos.iloc[:150,21:27]
# Se seleccionan las columnas que van a ser tenidas en cuenta para la comparacion
 # Todas son de Datos numericos
```

```python
jugadores
```

```python
jugadores = jugadores.values # convertir dataframe en array de arrays
```

```python
clasif = equipos.iloc[:150,[94]]# se selecciona la columna con las etiquetas de clasificacion
clasif = clasif.values                     
clasif
```

```python
clasif = clasif.reshape(-1)
```

```python
clasif
```

```python
# con la funcion train_test_split(,) se procede a realizar una division del dataset 
# ... entre datos de entrenamiento y datos de testing
X_train,X_test,y_train,y_test= train_test_split(jugadores,clasif) 
```

```python
red = MLPClassifier(max_iter=100000, hidden_layer_sizes=(4000,4000))
# maximo de 100000 con 4000 capas ocultas de 4000 nodos
```

```python
red.fit(X_train,y_train)# Se realiza el ajuste (entrenamiento) del algoritmo con los datos de entrenamiento
```

```python
red.score(X_test,y_test)# tasa de aprendizaje del algoritmo 
```

```python
x = equipos.iloc[150:,2] # se seleccionan los nmbres de los jugadores para la prediccion
x= x.values # # convertir dataframe en array de arrays
print(x)
```

```python
df_pred = equipos.iloc[150:,21:27] #lectura de los  registros para la nueva prediccion
```

```python
df_pred = df_pred.values # convertir dataframe en array de arrays 
```

```python
# tenemos las mediciones y queremos saber a que clasificacion pertenece
for i, valor in enumerate(df_pred):
    cont = equipos['y'][150+i]
    red.predict([valor])    # el tipo de i en estos momento es array
    x1 = x[i]
    clb = equipos['club'][150+i]
    
    if (red.predict([valor]) == 1):
        y = 'es de la liga española'
    else :
        y = 'es de la liga colombiana'
      
    if (int(red.predict([valor]))== (cont)):
        es = ',eso SI es cierto '
        
    else:
        es = ',eso NO es cierto'    
    
    print(x1, y, red.predict([valor]),es, 'ya que juega en',clb)
    print( '\n')
```
